### <m> Tensor Algebra using Numpy </m>

```python
B = np.arange(16,32).reshape(A.shape)

print(f">_ Addition of Tensors:\n {A + B}.\n")
print(f">_ Subtraction of Tensors:\n {A - B}.\n")
print(f">_ Hadamard Multiplication of Tensors:\n {A * B}.\n")
print(f">_ Simple division of Tensors:\n {A / B}.\n")
```

#### Tensor dot

```python
C = np.tensordot(A,B,axes=0)
print(f">_ Dot production of Tensors:\n {C}.\n")
print(f"The Shape of this production is: {C.shape}.")
```

### Read a RGB high resolution image

```python
Image = tl.tensor(plt.imread("./Images/orig.png"),dtype=np.float32)
print(f"The shape of image is {Image.shape}")

plt.imshow(Image)
plt.axis('off')
plt.show()
```

## CP decomposition

```python
Weights,Factors = tl.decomposition.parafac(Image,rank=100,init='random',tol=1e-8)
CP_Image = tl.cp_to_tensor((Weights,Factors))
```

### Tucker decomposition

```python
Core, TFactors = tl.decomposition.tucker(Image,rank=[100,100,3],init='random',tol=1e-8)
Tucker_Image = tl.tucker_to_tensor((Core,TFactors))
```

```python
fig = plt.figure(figsize=plt.figaspect(0.35))
ax = fig.add_subplot(1,3,1)
ax.imshow(Image)
ax.set_title("Original")
ax.axis('off')

ax2 = fig.add_subplot(1,3,2)
ax2.imshow(CP_Image)
ax2.set_title("CP decomposition")
ax2.axis('off')

ax3 = fig.add_subplot(1,3,3)
ax3.imshow(Tucker_Image)
ax3.set_title("Tucker decomposition")
ax3.axis('off')
```
